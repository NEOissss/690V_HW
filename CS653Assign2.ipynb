### For Experiment 1

```python
def exp1_draw(b, c, d, e):
    tcp_list = ["Tahoe", "Reno", "Newreno", "Vegas"]
    color = ["green","blue","yellow","red"] 
    dash = ['dashed','dotted','dotdash','dashdot']
    title = "CBR:"+str(b+1)+"-"+str(c + 11)+","+d+","+e
    plot1 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'latency(s)', plot_width=400, plot_height=400)
    plot2 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'throughput(bits)', plot_width=400, plot_height=400)
    plot3 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'drop rate', plot_width=400, plot_height=400)

    for i in range(len(tcp_list)):
        a = tcp_list[i]
        #b = -1
        #c = 1.0
        #d = "8Mb"
        #e = "L"
        filename = "DropTail"+"_"+"10Mb"+"_"+e+"_"+d+"_"+str(b+1)+"_"+str(c + 11)+"_"+"L"+"_"+a+"_"+"1"+"_"+"11"+".tr"
        re = analyze.analyze("exp1/"+filename,0.5)
        #print re
        plot1.line(re["datapoints"],re["N1-N4"]["latency"],line_width=2,color=color[i],legend=tcp_list[i],line_alpha=0.75,line_dash=dash[i])
        plot2.line(re["datapoints"],re["N1-N4"]["throughput"],line_width=2,color=color[i],legend=tcp_list[i],line_alpha=0.75,line_dash=dash[i])
        plot3.line(re["datapoints"],re["N1-N4"]["dropRate"],line_width=2,color=color[i],legend=tcp_list[i],line_alpha=0.75,line_dash=dash[i])

    plot1.legend.location = "bottom_right"
    plot2.legend.location = "bottom_right"
    plot3.legend.location = "bottom_right"
    show(column(row(plot1,plot2),plot3))
```

```python
# Parameters:
# exp1_draw(b, c, d, e)
# b >> gap_start = [-1, 0.1, 1.0, 4.0]
# c >> gap_stop = [1.0, -0.1, -1.0, -4.0]
# d >> cbr_bw = ["1Mb", "2Mb", "4Mb", "8Mb"]
# e >> cbr_dir = ["L", "R"]
```

### For Experiment 2

```python
def exp2_draw(i, c, d, e, f):
    tcp1_list = ["Reno", "Reno", "Vegas", "Vegas"]
    tcp2_list = ["Reno", "Newreno", "Vegas", "Newreno"]
    color = ["blue","red"] 
    title = "CBR:"+f+","+tcp1_list[i]+"&"+tcp2_list[i]+","+d+"&"+e
    plot1 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'latency(s)', plot_width=400, plot_height=400)
    plot2 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'throughput(bits)', plot_width=400, plot_height=400)
    plot3 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'drop rate', plot_width=400, plot_height=400)

    
    a = tcp1_list[i]
    b = tcp2_list[i]
    #c = -1
    #d = 1.0
    #e = "8Mb"
    #f = "L"
    filename = "DropTail"+"_"+"10Mb"+"_"+"L"+"_"+f+"_"+"12"+"_"+"22"+"_"+d+"_"+a+"_"+"10"+"_"+"25"+"_"+e+"_"+b+"_"+str(c+10)+"_"+"35"+".tr"
    re = analyze.analyze("exp2/"+filename,0.5)
    #print re
    plot1.line(re["datapoints"],re["N1-N4"]["latency"],line_width=2,color=color[0],legend=tcp1_list[i]+"("+d+")",line_alpha=0.6)
    plot1.line(re["datapoints"],re["N5-N6"]["latency"],line_width=2,color=color[1],legend=tcp2_list[i]+"("+e+")",line_alpha=0.6)
    plot2.line(re["datapoints"],re["N1-N4"]["throughput"],line_width=2,color=color[0],legend=tcp1_list[i]+"("+d+")",line_alpha=0.6)
    plot2.line(re["datapoints"],re["N5-N6"]["throughput"],line_width=2,color=color[1],legend=tcp2_list[i]+"("+e+")",line_alpha=0.6)
    plot3.line(re["datapoints"],re["N1-N4"]["dropRate"],line_width=2,color=color[0],legend=tcp1_list[i]+"("+d+")",line_alpha=0.6)
    plot3.line(re["datapoints"],re["N5-N6"]["dropRate"],line_width=2,color=color[1],legend=tcp2_list[i]+"("+e+")",line_alpha=0.6)
        
    plot1.legend.location = "bottom_right"
    plot2.legend.location = "bottom_right"
    plot3.legend.location = "bottom_right"
    show(column(row(plot1,plot2),plot3))
```

### For Experiment 3

```python
def exp3_draw(a, c):
    que_list = ["DropTail", "RED"]
    color = ["blue","red"] 
    title = a+","+c
    plot1 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'latency(s)', plot_width=400, plot_height=400)
    plot2 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'throughput(bits)', plot_width=400, plot_height=400)
    plot3 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'drop rate', plot_width=400, plot_height=400)

    for i in range(len(que_list)):
        #a = "Reno"
        b = que_list[i]
        #c = 1.0
        filename = b+"_"+"10Mb"+"_"+"L"+"_"+"8Mb"+"_"+"0"+"_"+"10"+"_"+c+"_"+a+"_"+"1"+"_"+"11"+".tr"
        re = analyze.analyze("exp3/"+filename,0.5)
        #print re
        plot1.line(re["datapoints"],re["N1-N4"]["latency"],line_width=2,color=color[i],legend=que_list[i],line_alpha=0.8)
        plot2.line(re["datapoints"],re["N1-N4"]["throughput"],line_width=2,color=color[i],legend=que_list[i],line_alpha=0.8)
        plot3.line(re["datapoints"],re["N1-N4"]["dropRate"],line_width=2,color=color[i],legend=que_list[i],line_alpha=0.8)

    plot1.legend.location = "bottom_right"
    plot2.legend.location = "bottom_right"
    plot3.legend.location = "bottom_right"
    show(column(row(plot1,plot2),plot3))
```
