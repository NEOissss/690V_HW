```python
def exp1_draw(a, b, c, e):
    #tcp_list = ["Tahoe", "Reno", "Newreno", "Vegas"]
    bw_list = ["1Mb", "2Mb", "4Mb", "8Mb"]
    color = ["green","blue","yellow","red"] 
    dash = ['dashed','dotted','dotdash','dashdot']
    title = a+",CBR:"+str(b+1)+"-"+str(c + 11)+","+e
    plot1 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'Latency (s)')
    plot2 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'Throughput (bytes)')
    plot3 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'Drops')
    plot4 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'SEQ Num')
    plot5 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'CWND')

    for i in range(len(bw_list)):
        #a = tcp_list[i]
        #b = -1
        #c = 1.0
        d = bw_list[i]
        #e = "L"
        filename = "DropTail"+"_"+"10Mb"+"_"+e+"_"+d+"_"+str(b+1)+"_"+str(c + 11)+"_"+"L"+"_"+a+"_"+"1"+"_"+"11"+".tr"
        re = analyze.analyze("exp1/"+filename,0.5)
        time = len(re["datapoints"])*0.5
        ave_lat = "%0.4f" % (sum(re["N1-N4"]["latency"])/10.0)
        ave_thr = "%0.1f" % (sum(re["N1-N4"]["throughput"])/10.0)
        plot1.line(re["datapoints"],re["N1-N4"]["latency"],line_width=2,color=color[i],legend=bw_list[i]+"(Avg:"+str(ave_lat)+")",line_alpha=0.7)
        plot2.line(re["datapoints"],re["N1-N4"]["throughput"],line_width=2,color=color[i],legend=bw_list[i]+"(Avg:"+str(ave_thr)+")",line_alpha=0.7)
        plot3.line(re["datapoints"],re["N1-N4"]["drops"],line_width=2,color=color[i],legend=bw_list[i],line_alpha=0.7)
        plot4.line(re["datapoints"],re["N1-N4"]["sequences"],line_width=2,color=color[i],legend=bw_list[i],line_alpha=0.7)
        plot5.line(re["datapoints"],re["N1-N4"]["windows"],line_width=2,color=color[i],legend=bw_list[i],line_alpha=0.7)
        
    plot1.legend.location = "bottom_right"
    plot2.legend.location = "bottom_right"
    plot3.legend.location = "top_right"
    plot4.legend.location = "top_left"
    plot5.legend.location = "bottom_right"
    show(row(plot1,plot2,plot3,plot4,plot5))
```

```python
# Average bar
tcp_list = ["Tahoe", "Reno", "Newreno", "Vegas"]
bw_list = ["1Mb", "2Mb", "4Mb", "8Mb"]
color = ["green","blue","yellow","red"] 
dash = ['dashed','dotted','dotdash','dashdot']

latency = {"bw":bw_list, "Tahoe":[], "Reno":[], "Newreno":[], "Vegas":[]}
throughput = {"bw":bw_list, "Tahoe":[], "Reno":[], "Newreno":[], "Vegas":[]}
drops = {"bw":bw_list, "Tahoe":[], "Reno":[], "Newreno":[], "Vegas":[]}
sequences = {"bw":bw_list, "Tahoe":[], "Reno":[], "Newreno":[], "Vegas":[]}
windows = {"bw":bw_list, "Tahoe":[], "Reno":[], "Newreno":[], "Vegas":[]}

for i in range(len(tcp_list)):
    for j in range(len(bw_list)):
        a = tcp_list[i]
        b = -1
        c = 1.0
        d = bw_list[j]
        e = "L"
        filename = "DropTail"+"_"+"10Mb"+"_"+e+"_"+d+"_"+str(b+1)+"_"+str(c + 11)+"_"+"L"+"_"+a+"_"+"1"+"_"+"11"+".tr"
        re = analyze.analyze("exp1/"+filename,0.5)
        time = len(re["datapoints"])*0.5
        ave_lat = sum(re["N1-N4"]["latency"])/10.0
        ave_thr = sum(re["N1-N4"]["throughput"])/10.0
        ave_drp = sum(re["N1-N4"]["drops"])/10.0
        lst_seq = max(re["N1-N4"]["sequences"])
        ave_wnd = sum(re["N1-N4"]["windows"])/10.0
        
        latency[tcp_list[i]].append(ave_lat)
        throughput[tcp_list[i]].append(ave_thr)
        drops[tcp_list[i]].append(ave_drp)
        sequences[tcp_list[i]].append(lst_seq)
        windows[tcp_list[i]].append(ave_wnd)
        
source1 = ColumnDataSource(latency)
source2 = ColumnDataSource(throughput)
source3 = ColumnDataSource(drops)
source4 = ColumnDataSource(sequences)
source5 = ColumnDataSource(windows)
```

```python
# this creates [ ("Apples", "2015"), ("Apples", "2016"), ("Apples", "2017"), ("Pears", "2015), ... ]
x = [ (bw, tcp) for bw in bw_list for tcp in tcp_list ]
counts1 = sum(zip(latency['Tahoe'], latency['Reno'], latency['Newreno'], latency['Vegas']), ()) # like an hstack
source1 = ColumnDataSource(data=dict(x=x, counts=counts1))
counts2 = sum(zip(throughput['Tahoe'], throughput['Reno'], throughput['Newreno'], throughput['Vegas']), ()) # like an hstack
source2 = ColumnDataSource(data=dict(x=x, counts=counts2))
counts3 = sum(zip(drops['Tahoe'], drops['Reno'], drops['Newreno'], drops['Vegas']), ()) # like an hstack
source3 = ColumnDataSource(data=dict(x=x, counts=counts3))
counts4 = sum(zip(sequences['Tahoe'], sequences['Reno'], sequences['Newreno'], sequences['Vegas']), ()) # like an hstack
source4 = ColumnDataSource(data=dict(x=x, counts=counts4))
counts5 = sum(zip(windows['Tahoe'], windows['Reno'], windows['Newreno'], windows['Vegas']), ()) # like an hstack
source5 = ColumnDataSource(data=dict(x=x, counts=counts5))

palette = ["#c9d9d3", "#718dbf", "#e84d60", "#c27ba0"]

p1 = figure(x_range=FactorRange(*x), plot_height=250, y_range=(0.06,0.072), title="Average Latency (s)", toolbar_location=None, tools="")
p1.vbar(x='x', top='counts', width=0.9, source=source1, line_color="white", fill_color=factor_cmap('x', palette=palette, factors=tcp_list, start=1, end=2))
p1.y_range.start = 0
p1.x_range.range_padding = 0.1
p1.xaxis.major_label_orientation = 1
p1.xgrid.grid_line_color = None

p2 = figure(x_range=FactorRange(*x), plot_height=250, title="Average Throughput (bytes)", toolbar_location=None, tools="")
p2.vbar(x='x', top='counts', width=0.9, source=source2, line_color="white", fill_color=factor_cmap('x', palette=palette, factors=tcp_list, start=1, end=2))
p2.y_range.start = 0
p2.x_range.range_padding = 0.1
p2.xaxis.major_label_orientation = 1
p2.xgrid.grid_line_color = None

p3 = figure(x_range=FactorRange(*x), plot_height=250, title="Average Drops", toolbar_location=None, tools="")
p3.vbar(x='x', top='counts', width=0.9, source=source3, line_color="white", fill_color=factor_cmap('x', palette=palette, factors=tcp_list, start=1, end=2))
p3.y_range.start = 0
p3.x_range.range_padding = 0.1
p3.xaxis.major_label_orientation = 1
p3.xgrid.grid_line_color = None

p4 = figure(x_range=FactorRange(*x), plot_height=250, title="Total SEQ Num", toolbar_location=None, tools="")
p4.vbar(x='x', top='counts', width=0.9, source=source4, line_color="white", fill_color=factor_cmap('x', palette=palette, factors=tcp_list, start=1, end=2))
p4.y_range.start = 0
p4.x_range.range_padding = 0.1
p4.xaxis.major_label_orientation = 1
p4.xgrid.grid_line_color = None

p5 = figure(x_range=FactorRange(*x), plot_height=250, title="Average CWND", toolbar_location=None, tools="")
p5.vbar(x='x', top='counts', width=0.9, source=source5, line_color="white", fill_color=factor_cmap('x', palette=palette, factors=tcp_list, start=1, end=2))
p5.y_range.start = 0
p5.x_range.range_padding = 0.1
p5.xaxis.major_label_orientation = 1
p5.xgrid.grid_line_color = None

show(row(p1,p2,p3,p4,p5))
```

```python
def exp2_draw(i, c, d, e, f):
    tcp1_list = ["Reno", "Reno", "Vegas", "Vegas"]
    tcp2_list = ["Reno", "Newreno", "Vegas", "Newreno"]
    color = ["blue","red"] 
    title = "CBR:"+f+","+tcp1_list[i]+"&"+tcp2_list[i]+","+d+"&"+e
    plot1 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'Latency(s)')
    plot2 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'Throughput(bytes)')
    plot3 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'Drop rate')
    plot4 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'SEQ Num')
    plot5 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'CWND')
    
    a = tcp1_list[i]
    b = tcp2_list[i]
    #c = -1
    #d = 1.0
    #e = "8Mb"
    #f = "L"
    filename = "DropTail"+"_"+"10Mb"+"_"+"L"+"_"+f+"_"+"12"+"_"+"22"+"_"+d+"_"+a+"_"+"10"+"_"+"25"+"_"+e+"_"+b+"_"+str(c+10)+"_"+"25"+".tr"
    re = analyze.analyze("exp2/"+filename,0.5)
    time = len(re["datapoints"])*0.5
    ave_lat1 = "%.4f" % (sum(re["N1-N4"]["latency"])/time)
    ave_thr1 = sum(re["N1-N4"]["throughput"])/time
    ave_lat2 = sum(re["N5-N6"]["latency"])/time
    ave_thr2 = sum(re["N5-N6"]["throughput"])/time
    #print re
    plot1.line(re["datapoints"],re["N1-N4"]["latency"],line_width=2,color=color[0],legend=tcp1_list[i]+"("+d+")"+"(Avg:"+str(ave_lat1)+")",line_alpha=0.7)
    plot1.line(re["datapoints"],re["N5-N6"]["latency"],line_width=2,color=color[1],legend=tcp2_list[i]+"("+e+")"+"(Avg:"+str(ave_lat2)+")",line_alpha=0.7)
    plot2.line(re["datapoints"],re["N1-N4"]["throughput"],line_width=2,color=color[0],legend=tcp1_list[i]+"("+d+")"+"(Avg:"+str(ave_thr1)+")",line_alpha=0.7)
    plot2.line(re["datapoints"],re["N5-N6"]["throughput"],line_width=2,color=color[1],legend=tcp2_list[i]+"("+e+")"+"(Avg:"+str(ave_thr2)+")",line_alpha=0.7)
    plot3.line(re["datapoints"],re["N1-N4"]["dropRate"],line_width=2,color=color[0],legend=tcp1_list[i]+"("+d+")",line_alpha=0.7)
    plot3.line(re["datapoints"],re["N5-N6"]["dropRate"],line_width=2,color=color[1],legend=tcp2_list[i]+"("+e+")",line_alpha=0.7)
    plot4.line(re["datapoints"],re["N1-N4"]["sequences"],line_width=2,color=color[0],legend=tcp1_list[i]+"("+d+")",line_alpha=0.7)
    plot4.line(re["datapoints"],re["N5-N6"]["sequences"],line_width=2,color=color[1],legend=tcp2_list[i]+"("+e+")",line_alpha=0.7)
    plot5.line(re["datapoints"],re["N1-N4"]["windows"],line_width=2,color=color[0],legend=tcp1_list[i]+"("+d+")",line_alpha=0.7)
    plot5.line(re["datapoints"],re["N5-N6"]["windows"],line_width=2,color=color[1],legend=tcp2_list[i]+"("+e+")",line_alpha=0.7)

   
    plot1.legend.location = "bottom_right"
    plot2.legend.location = "bottom_right"
    plot3.legend.location = "bottom_right"
    plot4.legend.location = "top_left"
    plot5.legend.location = "bottom_right"
    show(row(plot1,plot2,plot3,plot4,plot5))
```

```python
def exp3_draw(a, c):
    que_list = ["DropTail", "RED"]
    color = ["blue","red"] 
    title = a+","+c
    plot1 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'Latency(s)')
    plot2 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'Throughput(bytes)')
    plot3 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'Drop rate')
    plot4 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'SEQ Num')
    plot5 = figure(title = title, x_axis_label = 'time(s)', y_axis_label = 'CWND')
    
    for i in range(len(que_list)):
        #a = "Reno"
        b = que_list[i]
        #c = 1.0
        filename = b+"_"+"10Mb"+"_"+"L"+"_"+"4Mb"+"_"+"2"+"_"+"10"+"_"+c+"_"+a+"_"+"0"+"_"+"12"+".tr"
        re = analyze.analyze("exp3/"+filename,0.5)
        time = len(re["datapoints"])*0.5
        ave_lat = sum(re["N1-N4"]["latency"])/time
        ave_thr = sum(re["N1-N4"]["throughput"])/time
        
        plot1.line(re["datapoints"],re["N1-N4"]["latency"],line_width=2,color=color[i],legend=que_list[i]+"(Avg:"+str(ave_lat)+")",line_alpha=0.7)
        plot2.line(re["datapoints"],re["N1-N4"]["throughput"],line_width=2,color=color[i],legend=que_list[i]+"(Avg:"+str(ave_thr)+")",line_alpha=0.7)
        plot3.line(re["datapoints"],re["N1-N4"]["dropRate"],line_width=2,color=color[i],legend=que_list[i],line_alpha=0.7)
        plot4.line(re["datapoints"],re["N1-N4"]["sequences"],line_width=2,color=color[i],legend=que_list[i],line_alpha=0.7)
        plot5.line(re["datapoints"],re["N1-N4"]["windows"],line_width=2,color=color[i],legend=que_list[i],line_alpha=0.7)

    plot1.legend.location = "bottom_right"
    plot2.legend.location = "bottom_right"
    plot3.legend.location = "bottom_right"
    plot4.legend.location = "top_left"
    plot5.legend.location = "bottom_right"
    show(row(plot1,plot2,plot3,plot4,plot5))
```
